```python
# what if you wanna use a custom function to group by 

#based on temperature 
# idx = row , col = col 

def grouper(df,idx,col):
    if 80<=df[col].loc[idx] <=90:
        return '80-90'
    elif 50<=df[col].loc[idx] <=60:
        return '50-60'
    else:
        return 'others'
```

```python
# create a group object based on custom column
g = df.groupby(lambda idx: grouper(df,idx,'temperature'))
g

```

```python
for key, data in g:
    print('key',key)
    print('data', data)
```

# Concat and Merge 

```python
data = {
    'city': ['Delhi', 'Mumbai', 'Bangalore'],
    'temp': [30, 28, 25],
    'humidity': [60, 75, 50]
}

india_weather = pd.DataFrame(data)
india_weather
```

```python
data = {
    'city': ['New York', 'Los Angeles', 'Chicago'],
    'temp': [25, 30, 20],
    'humidity': [50, 65, 40]
}

us_weather = pd.DataFrame(data)
us_weather
```

```python
#merge these two data frames

df = pd.concat([india_weather,us_weather])
df

# index remains from the original dataframe
```

```python
df = pd.concat([india_weather,us_weather], ignore_index= True)
df
```

```python
# create keys 

df = pd.concat([india_weather,us_weather], keys = ['India','US'])
df
```

```python
df.loc['India']
```
